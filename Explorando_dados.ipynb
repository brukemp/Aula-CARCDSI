```python
import pandas as pd
```

```python
dados = {"sexo": ["F", "M", "F", "F", "F", "M"],
         "bairro": ["Belverde", "Belverde", "Savassi", "Anchieta", None, "Savassi"],
         "valor": [150.00, 35.00, 80.00, 250.00, 9.90, 25.00],
         "cartao": ["Master", "Visa", "Visa", "Anex", "Elo", "Master"]}
id_clientes = [1, 2, 3, 4, 5, 6]

vendas = pd.DataFrame(dados, index=id_clientes)
```

```python
print("Dominio dos atributos categóricos")
print("---------------------------------")
print("sexo: ", vendas["sexo"].unique())
print("bairro: ", vendas["bairro"].unique())
print("cartao: ", vendas["cartao"].unique())
```

```python
print("Tabela de frequência")
print("--------------------")
print("\n1 - sexo")
print(vendas["sexo"].value_counts())
print("\n2 - bairro")
print(vendas["bairro"].value_counts())
print("\n3 - cartão")
print(vendas["cartao"].value_counts())
```

```python
grupo_valor_bairro = vendas['valor'].groupby(vendas['bairro'])

grupo_valor_bairro1 = vendas.groupby("bairro")["valor"]

print(grupo_valor_bairro.count())
print(grupo_valor_bairro.sum())
print(grupo_valor_bairro.mean())
```

```python
# Estudando a base de dados Flags
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

# importaando os dados
flags = pd.read_csv("/content/drive/MyDrive/CARCDSI/MaterialApoio/flags.csv")

flags
```

```python
# Obtendo as propriedades básicas de cada atributo
i = 0
for c in flags.columns:
  i += 1
  att = flags [c]                       # atributp
  att_dtype = att.dtype                 # dtype
  att_tam_dominio = att.unique().size   # tamanho do dominio
  att_tem_nulo = any(att. isnull())
  if(att_tam_dominio < 8):
    print("("+str(i)+") atributo: ", c, "\t",
          "dtype: ", att_dtype, "\t",
          "nulos: ", att_tem_nulo, "\n",
          "dominio: ", att.unique())
  else:
    if(att_dtype == 'object'):
      print("("+str(i)+") atributo: ", c, "\t",
            "dtype: ", att_dtype, "\t",
            "nulos: ", att_tem_nulo, "\n",
            "dominio (primeiros elementos)", att.unique()[0:8])
    else:
      print("("+str(i)+") atributo: ", c, "\t",
            "dtype: ", att_dtype, "\t",
            "nulos: ", att_tem_nulo, "\n",
            "min: ", att.min(), "\t"
            "max: ", att.max(), "\t"
            "media: ", round(att.mean(),2), "\t"
            "d.p,: ", round(att.std(),2))
```

```python
df_cores = pd.DataFrame()
for c in flags.columns:
  if c in ["red", "green", "blue", "gold", "white", "black", "orange"]:
    df_cores[c] = flags[c].value_counts()

print(df_cores)
```

```python
lst_cores = ["red", "green", "blue", "gold", "white", "black", "orange"]

df_cores.plot(kind="barh", subplots=True, figsize=(8, 25), color=lst_cores)
```
